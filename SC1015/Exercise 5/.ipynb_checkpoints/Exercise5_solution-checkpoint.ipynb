```python
from pandas.api.types import is_numeric_dtype
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import OrdinalEncoder
from sklearn.tree import plot_tree

import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df_main = pd.read_csv("./train.csv")
```

## Problem 1 : Predicting CentralAir using SalePrice

a) Create appropriate datasets for Train and Test in an 75:25 ratio and fit two Classification Tree models (of max
depth 3 and max depth 4) on the Train set to predict CentralAir using SalePrice. Print the tree in each case.

```python
class UnivariateModel:
    def __init__(self, feature, label, max_depth):
        self.feature = feature
        self.label = label
        self.df_train = df_main[[feature, label]].copy()
        self.encoder = OrdinalEncoder()
        self.max_depth = max_depth

#         self.encode_data()
        self.train_model()
        
    def encode_data(self):
        for col in list(self.df_train.columns):
            if not is_numeric_dtype(self.df_train[col]):
                self.encoder.fit(self.df_train[col].to_frame())
                self.df_train[col] = self.encoder.transform(self.df_train[col].to_frame())

    def train_model(self):
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(
            self.df_train[self.feature].to_frame(),
            self.df_train[self.label].to_frame(),
            test_size=0.25
        )

        self.dectree = DecisionTreeClassifier(max_depth=self.max_depth)
        self.dectree = self.dectree.fit(self.X_train, self.y_train)
        
        self.y_train_pred = self.dectree.predict(self.X_train)
        self.y_test_pred = self.dectree.predict(self.X_test)
        
    def get_goodness_fit(self):
        train_cm = confusion_matrix(self.y_train, self.y_train_pred)
        print("Goodness of Fit of Model \tTrain Dataset")
        print("Classification Accuracy \t:", self.dectree.score(self.X_train, self.y_train))
        print("True Positive Rate (TP/TP+FN) \t:", train_cm[1][1]/(np.sum(train_cm[1])))
        print("False Positive Rate (FP/TN+FP) \t:", train_cm[1][0]/(np.sum(train_cm[0])))
        print()

        test_cm = confusion_matrix(self.y_test, self.y_test_pred)
        print("Goodness of Fit of Model \tTest Dataset")
        print("Classification Accuracy \t:", self.dectree.score(self.X_test, self.y_test))
        print("True Positive Rate (TP/TP+FN) \t:", test_cm[1][1]/(np.sum(test_cm[1])))
        print("False Positive Rate (FP/TN+FP) \t:", test_cm[1][0]/(np.sum(test_cm[0])))
        print()
        
    def plot_dectree(self):
        f, ax = plt.subplots(figsize=(5*self.max_depth,5*self.max_depth))
        plot_tree(self.dectree, max_depth=self.max_depth, filled=True, rounded=True, ax=ax,
                  feature_names=[self.feature], class_names=self.dectree.classes_)
#         class_names=list(df_main[self.label].unique())
        f.suptitle(f"Decision tree of {self.label} classified by {self.feature} (max depth = {self.max_depth})", y=0.85)
        plt.show()
        
    def show_confusion_matrix(self):
        f, axes = plt.subplots(1, 2, figsize=(12, 4))
        sb.heatmap(confusion_matrix(self.y_train, self.y_train_pred),
                   annot=True, fmt=".0f", annot_kws={"size": 18}, ax=axes[0])
        sb.heatmap(confusion_matrix(self.y_test, self.y_test_pred), 
                   annot=True, fmt=".0f", annot_kws={"size": 18}, ax=axes[1])
        f.suptitle(f"Confusion matrix of {self.label} classified by {self.feature} (max depth = {self.max_depth})")
        plt.show()
```

```python
model_SalePrice_3 = UnivariateModel('SalePrice', 'CentralAir', 3)
model_SalePrice_4 = UnivariateModel('SalePrice', 'CentralAir', 4)
```

```python
# Max depth 3
model_SalePrice_3.plot_dectree()
```

```python
# Max depth 4
model_SalePrice_4.plot_dectree()
```

b) Print the confusion matrix on Train set and Test set for both the aforesaid models. Compute and print for both
the trees the Classification Accuracy, True Positive Rate, False Positive Rate. Which of the two models is better?

```python
for idx, model in enumerate((model_SalePrice_3, model_SalePrice_4)):
    model.show_confusion_matrix()
    model.get_goodness_fit()
```

<b>Conclusion:</b> On average, although both have similar classification accuracy, the model with max depth of 4 is better as it has a lower FPR and higher TPR than the one with a max depth of 3.

Following the steps from the previous problem, build two new uni-variate Classification Tree models (of max depth 3) to

represents a specific partition of your training data, based on certain conditions given by the splits in the tree.
Find out which leaf node of this tree has the highest number of False Positives (FP) in the training dataset and
print all samples in the Train set (rows from the dataframe) that end up in this leaf node (partition) during fit.

```python
model_SalePrice_3.plot_dectree()
```

```python
model_SalePrice_3.dectree.classes_
```

<b>Conclusion:</b> Third leaf node from the left has the highest number of false positives (i.e. predicted Y or 1 but classified as N or 0) at 26.

```python
df_train_SalePrice_3 = model_SalePrice_3.df_train.loc[
    model_SalePrice_3.df_train.index.isin(model_SalePrice_3.y_train.index)]

# Print all samples in that leaf node
all_in_leaf = df_train_SalePrice_3[
    (df_train_SalePrice_3['SalePrice'] > 79250) &
    (df_train_SalePrice_3['SalePrice'] <= 105950)
]
print("All samples in leaf:")
print(all_in_leaf)
print()
print(f"Number of samples: {len(all_in_leaf)}")
print()

# Print false positives only
fp = df_train_SalePrice_3[
    (df_train_SalePrice_3['SalePrice'] > 79250) &
    (df_train_SalePrice_3['SalePrice'] <= 105950) &
    (df_train_SalePrice_3['CentralAir'] == 'N')
]
print("False positives in leaf:")
print(fp)
print()
print(f"Number of samples: {len(fp)}")
```
